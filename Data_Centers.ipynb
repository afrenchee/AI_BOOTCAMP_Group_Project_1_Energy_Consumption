```python
import matplotlib.pyplot as plt

# Data for the graph
data_centers = ["Switch Citadel", "Google (The Dalles)", "Microsoft (Quincy)"]
data_center_consumption = [7.45, 12, 5.26]  # TWh

states = ["Vermont", "Wyoming", "Rhode Island"]
state_consumption = [5.97, 14.7, 7.6]  # TWh

# Plotting
plt.figure(figsize=(10, 6))

# Plotting data centers
plt.bar(data_centers, data_center_consumption, label='Data Centers', alpha=0.7)

# Plotting states
plt.bar(states, state_consumption, label='States', alpha=0.7)

# Adding labels and title
plt.xlabel('Data Centers and States')
plt.ylabel('Annual Energy Consumption (TWh)')
plt.title('Annual Energy Consumption Comparison: Data Centers vs Small States')
plt.xticks(rotation=45, ha='right')
# Adding legend
plt.legend()

# Display the plot
plt.show()
```
